#**Modelo**


##¿Como represento el espacio de soluciones?

El número mínimo de días de grabación está determinado por la restricción del problema:

30 tomas / 6 tomas por día = 5 días mínimo.

Si aplicamos combinatoria para saber cuántas formas distintas existen de repartir las 30 tomas en esos 5 días, tenemos:

      Elegir 6 tomas para el Día 1: C(30,6) =    593.775
      Elegir 6 tomas para el Día 2: C(24,6) =    134.596
      Elegir 6 tomas para el Día 3: C(18,6) =     18.564
      Elegir 6 tomas para el Día 4: C(12,6) =        924
      Elegir 6 tomas para el Día 5: C(6,6)  =          1

      Total solo para 5 días exactos:593.775 × 134.596 × 18.564 × 924 × 1 ≈ 1.37 × 10^18 combinaciones.

Este cálculo representa solo el caso de 5 días exactos. Si se tienen en cuenta otra soluciones con más días, el espacio a considerar será mayor.

El espacio de soluciones de este problema supera las 10^18 combinaciones posibles. Asumiendo una evaluación de 10^9 soluciones por segundo (estimación estandar basada en procesadores modernos de ~3GHz), el tiempo necesario para explorar todo el espacio seria ≈  31 años ≈ 10^18 / 10^9 ≈ 10^9 segundos.

Por lo cual usar un algoritmo como fuerza bruta se hace inviable, y se justifica el uso de una metaheurística.



##¿Cual es la función objetivo?

La función objetivo mide el costo total de convocar actores a lo largo de todos los días de grabación. Dado que cada actor cobra una tarifa fija por día que asiste al estudio, independientemente del número de toma que grabe ese día, el costo de un día equivale al número de actores distintos convocados ese día.

| Toma | A1 | A2 | A3 | A4 | A5 | A6 | A7 | A8 | A9 | A10 | Total |
|------|----|----|----|----|----|----|----|----|----|----|-------|

      - Actores T1 = {1,2,3,4,5}
      - Actores T2 = {3,4,5}
      - Actores T3 = {2,5,7}

      Unión    = {1,2,3,4,5,7}
      Costo    = 6  ->  6 actores para 3 tomas

      S = Solución (vector de asignación)
      d = Cada día de grabación

      T_d = Conjunto de tomas asignadas al día d
      A(t) = Conjunto de actores que participan en la toma t
      U = Unión de conjuntos (sin repetir)

      - El costo de un día d es:

          Costo(d) = |U A(t)| Para todo t ∈ T_d  (Representa la unión de todas las tomas del día)

      - La función objetivo a minimizar es:
         
         F(S) = ∑ Costo(d)  Para todo d ∈ D
              = ∑  |u A(t)|   






##¿Cómo implemento las restricciones?

**Restricción** : No es posible grabar más de 6 tomas por día

      Para todo día d: |T_d| ≤ 6

      T_d = Conjunto de tomas asignadas al día d    

Este problema representa una restricción dura, por lo cual ninguna solución valida puede violarla bajo ninguna circunstancia.

Para garantizar que el Algoritmo genético trabaje siempre con soluciones validas, se aplican dos estrategias complementarias:

 1. Construcción válida: Generando soluciones iniciales que ya cumplan la restricción desde el principio.  Por lo cual al ir asignando tomas, se verificará que el día destino no supere las 6 tomas.
    * Si día d tiene menos de 6 tomas -> Se asigna
    * Si día d tiene 6 tomas -> Busca otro día
    * Si todos los días están llenos -> Crea un día nuevo
 2. Reparación: Permite que la mutación genere soluciones inválidas, para repararlas inmediatamente despues.

    Despues del cruce: Día 3 tiene 8 tomas  (viola la restricción)
    * Toma 2 tomas del día 3
    * Mover a otro día con espacio disponible
    * Solución reparada

Utilizando estas dos estrategías se garantiza que el algortimo nunca evalúe soluciones inválidas.

# Función de costo (Fitness)

Implementa la funcion objetivo:

      F(S) = Σ | ∪ A(t) | para todo día d

      - Recorre cada toma y la agrupa por día
      - Para cada día calcula los actores únicos convocados
      - Retorna la suma total (la cual queremos minimizar)




```
def calcular_costo(solucion):
    """
    Recibe un vector de 30 enteros donde solucion[i] = día asignado a la toma i.
    Retorna el costo total = suma de actores únicos convocados por día.
    """
    dias = {} # Diccionario: día -> conjunto de actores
    for toma, dia in enumerate(solucion):
        if dia not in dias:
            dias[dia] = set()
        for actor in range(N_ACTORES):
            if M[toma][actor] == 1:
                dias[dia].add(actor)

    costo_total = sum(len(actores) for actores in dias.values())
    return costo_total

def es_valida(solucion):
    """
    Verifica que ningún día tenga más de MAX_TOMAS_DIA tomas (6)
    """
    conteo = {}
    for dia in solucion:
        conteo[dia] = conteo.get(dia, 0) + 1
    return all(v <= MAX_TOMAS_DIA for v in conteo.values())
```

# Generación de población inicial

Para que el GA tenga de dónde partir, se genera una población inicial de P soluciones aleatorias válidas

      Cada solución se construye asignando las 30 tomas a días
      de forma aleatoria, respetando desde el principio la
      restricción de máximo 6 tomas por día:

      - Si el día tiene menos de 6 tomas → se asigna
      - Si el día tiene 6 tomas → busca otro día
      - Si todos los días están llenos → crea un día nuevo

      Esto garantiza que todas las soluciones iniciales sean válidas antes de entrar al GA.

```
def generar_solucion():
    """
    Genera una asignación aleatoria válida de tomas a días.
    Se garantiza que ningún día supere MAX_TOMAS_DIA tomas.
    Retorna: Vector de 30 enteros

    """
    solucion = [0] * N_TOMAS
    capacidad = {}  # día -> tomas asignadas

    tomas_shuffled = list(range(N_TOMAS))
    random.shuffle(tomas_shuffled)

    dia_actual = 1
    for toma in tomas_shuffled:
        # Buscar un día con espacio disponible
        asignado = False
        dias_disponibles = [d for d, c in capacidad.items() if c < MAX_TOMAS_DIA]
        random.shuffle(dias_disponibles)

        for dia in dias_disponibles:
            solucion[toma] = dia
            capacidad[dia] = capacidad.get(dia, 0) + 1
            asignado = True
            break

        if not asignado:
            # Crear nuevo día
            solucion[toma] = dia_actual
            capacidad[dia_actual] = 1
            dia_actual += 1

    return solucion

def generar_poblacion(tam_poblacion):
    return [generar_solucion() for _ in range(tam_poblacion)]
```

```
#Test

sol = generar_solucion()

conteo = {}
for dia in sol:
    conteo[dia] = conteo.get(dia, 0) + 1

print(f"Solución generada:  {sol}")
print(f"Tomas por día:      {dict(sorted(conteo.items()))}")
print(f"Total tomas:        {sum(conteo.values())}")
print(f"¿Válida?:           {es_valida(sol)}")
print(f"Costo:              {calcular_costo(sol)}")

poblacion = generar_poblacion(100)
invalidas = sum(1 for s in poblacion if not es_valida(s))
print(f"\nPoblación de 100 → Inválidas: {invalidas}")
print(f"Costo promedio: {sum(calcular_costo(s) for s in poblacion)/len(poblacion):.1f}")
```

# Selección por torneo

La selección determina qué individuos de la población tienen derecho a reproducirse para generar la siguiente generación.

      Se utiliza selección por torneo:
      1. Se eligen k individuos al azar de la población
      2. Se comparan sus costos
      3. Gana el de menor costo → se convierte en padre

      Lo cual garantiza que los mejores individuos tengan más probabilidad de reproducirse, sin descartar completamente a los peores (mantiene diversidad)



```
def seleccion_torneo(poblacion, k=3):
    """
    Selecciona k individuos al azar y retorna el mejor (menor costo).
    """
    competidores = random.sample(poblacion, k)
    return min(competidores, key=calcular_costo)
```

```
#Test
poblacion = generar_poblacion(10)

# Mostrar costos de toda la población
costos = [calcular_costo(ind) for ind in poblacion]
print(f"Costos de la población: {costos}")
print(f"Mejor costo disponible: {min(costos)}")

# Seleccionar un ganador por torneo
ganador = seleccion_torneo(poblacion)
print(f"Costo del ganador:      {calcular_costo(ganador)}")
```

En esta fase a pesar que el ganador tiene costo de 43, (no 41 que es el minímo global), debido a que el torneo selecciona K = 3 individuos al azar y elige el mejor entre ellos. Manteniendo la diversidad de la población y evitando que el GA converja hacia una única solución.

# Cruce

El cruce combina dos soluciones padre para generar un hijo que hereda características de ambos.

      Se utiliza cruce de un punto:
        1. Se elige un punto de corte aleatorio
        2. El hijo toma la primera parte del Padre 1
        3. Y la segunda parte del Padre 2
        4. Si el hijo viola la restricción (se repara)

```
def cruce(padre1, padre2):
    """
    Cruce de un punto: combina dos soluciones y repara si viola restricciones.
    """
    punto = random.randint(1, N_TOMAS - 1)
    hijo = padre1[:punto] + padre2[punto:]
    hijo = reparar(hijo)
    return hijo

def reparar(solucion):
    """
    Si algún día supera MAX_TOMAS_DIA tomas, mueve el exceso a otro día.
    """
    solucion = list(solucion)
    conteo = {}
    for dia in solucion:
        conteo[dia] = conteo.get(dia, 0) + 1

    dia_nuevo = max(conteo.keys()) + 1

    for i in range(N_TOMAS):
        dia = solucion[i]
        if conteo[dia] > MAX_TOMAS_DIA:
            # Buscar día con espacio
            movido = False
            for d in sorted(conteo.keys()):
                if d != dia and conteo[d] < MAX_TOMAS_DIA:
                    conteo[dia] -= 1
                    solucion[i] = d
                    conteo[d] += 1
                    movido = True
                    break
            if not movido:
                # Crear día nuevo
                conteo[dia] -= 1
                solucion[i] = dia_nuevo
                conteo[dia_nuevo] = 1
                dia_nuevo += 1

    return solucion
```

La solución resultante es válida, lo que confirma que la función reparar() funciona correctamente. A lo largo de las generaciones, el GA irá encontrando combinaciones cada vez mejores.

# Mutación

La mutación introduce variabilidad aleatoria en las soluciones, evitando que toda la población converja hacia el mismo punto.

        Con una probabilidad se modifica una solución:
        1. Se elige una toma aleatoria
        2. Se mueve a un día diferente con espacio disponible

```
def mutar(solucion, prob_mutacion=0.1):
    solucion = list(solucion)

    if random.random() < prob_mutacion:
        toma   = random.randint(0, N_TOMAS - 1)

        conteo = {}
        for dia in solucion:
            conteo[dia] = conteo.get(dia, 0) + 1

        dias_disponibles = [d for d, c in conteo.items()
                            if c < MAX_TOMAS_DIA]

        # Si no hay días disponibles → crear uno nuevo
        if not dias_disponibles:
            dia_nuevo = max(conteo.keys()) + 1
            solucion[toma] = dia_nuevo
        else:
            solucion[toma] = random.choice(dias_disponibles)

    return solucion
```

```
# Test

sol_original = generar_solucion()
sol_mutada   = mutar(sol_original, prob_mutacion=1.0)

print(f"Original: {sol_original}")
print(f"Mutada:   {sol_mutada}")
print()

# Encontrar qué posición cambió
cambios = [i for i in range(N_TOMAS)
           if sol_original[i] != sol_mutada[i]]
print(f"Posición que cambió (toma): {cambios}")
for i in cambios:
    print(f"  Toma {i+1}: Día {sol_original[i]} → Día {sol_mutada[i]}")

print()
print(f"Costo original: {calcular_costo(sol_original)}")
print(f"Costo mutada:   {calcular_costo(sol_mutada)}")
print(f"¿Mutada válida?: {es_valida(sol_mutada)}")
```

La mutación movió la Toma 29 del Día 1 al Día 6 (nuevo). El costo empeoró de 46 a 48 porque la toma quedó sola en un día nuevo, sin compartir actores con otras tomas.

La mutación no busca mejorar directamente, sino introducir variabilidad para que el GA explore nuevas regiones del espacio de búsqueda. El GA buscará seleccionar las mejores soluciones en cada generación.

# Búsqueda local

La búsqueda local intenta mejorar una solución intercambiando los días de dos tomas aleatorias:

      1. Se eligen dos tomas al azar
      2. Se intercambian sus días asignados
      3. Si el nuevo costo es menor se acepta
      4. Si no mejora se descarta
      5. Se repite  n intentos veces

```
def busqueda_local(solucion, intentos=20):
    """
    Intenta mejorar la solución intercambiando tomas entre días.
    Inspirado en el pairwise interchange de Nordström & Tufekci (1994).
    """
    mejor = list(solucion)
    mejor_costo = calcular_costo(mejor)

    for _ in range(intentos):
        # Elegir dos tomas al azar e intercambiar sus días
        i, j = random.sample(range(N_TOMAS), 2)
        vecino = list(mejor)
        vecino[i], vecino[j] = vecino[j], vecino[i]

        costo_vecino = calcular_costo(vecino)
        if costo_vecino < mejor_costo:
            mejor = vecino
            mejor_costo = costo_vecino

    return mejor
```

```
#Tests
sol          = generar_solucion()
sol_mejorada = busqueda_local(sol, intentos=20)

print(f"Antes  → Costo: {calcular_costo(sol)}")
print(f"Después → Costo: {calcular_costo(sol_mejorada)}")
print(f"Mejora: {calcular_costo(sol) - calcular_costo(sol_mejorada)} unidades")
print(f"¿Válida?: {es_valida(sol_mejorada)}")

sol_mejorada_100 = busqueda_local(sol, intentos=100)
print(f"\nCon 100 intentos → Costo: {calcular_costo(sol_mejorada_100)}")
print(f"Mejora: {calcular_costo(sol) - calcular_costo(sol_mejorada_100)} unidades")
```

Podemos ver que con 20 intentos de intercambio, la búsqueda local logró reducir el costo de 46 a 38, una mejora de 8 unidades sin necesidad de generar una nueva solución desde cero, cuando se hace con 100 intentos se mantuvo en 38, indicando que con 20 intentos ya se encontró un minímo local.

# Pipeline de Algoritmo Genético completo

Una vez implemenatados todos los operadores, se integran  en un ciclo evolutivo que se repite durante 500 generaciones, hasta encontrar la planificación de menor costo.

```
def algoritmo_genetico(
    tam_poblacion=100,
    generaciones=500,
    prob_cruce=0.8,
    prob_mutacion=0.1,
    elitismo=2,
    usar_busqueda_local=True
):
    # --- Inicialización ---
    poblacion = generar_poblacion(tam_poblacion)
    historial_mejor = []
    historial_promedio = []

    mejor_global = min(poblacion, key=calcular_costo)
    mejor_costo_global = calcular_costo(mejor_global)

    print(f"Costo inicial (mejor de población): {mejor_costo_global}")

    for gen in range(generaciones):
        nueva_poblacion = []

        # --- Elitismo: conservar los mejores ---
        poblacion_ordenada = sorted(poblacion, key=calcular_costo)
        nueva_poblacion.extend(poblacion_ordenada[:elitismo])

        # --- Generar el resto de la nueva población ---
        while len(nueva_poblacion) < tam_poblacion:
            # Selección
            padre1 = seleccion_torneo(poblacion)
            padre2 = seleccion_torneo(poblacion)

            # Cruce
            if random.random() < prob_cruce:
                hijo = cruce(padre1, padre2)
            else:
                hijo = list(padre1)

            # Mutación
            hijo = mutar(hijo, prob_mutacion)

            # Búsqueda local (mejora memética)
            if usar_busqueda_local:
                hijo = busqueda_local(hijo)

            nueva_poblacion.append(hijo)

        poblacion = nueva_poblacion

        # --- Registrar estadísticas ---
        costos = [calcular_costo(ind) for ind in poblacion]
        mejor_gen = min(costos)
        prom_gen  = sum(costos) / len(costos)
        historial_mejor.append(mejor_gen)
        historial_promedio.append(prom_gen)

        if mejor_gen < mejor_costo_global:
            mejor_costo_global = mejor_gen
            mejor_global = poblacion[costos.index(mejor_gen)]

        # Imprimir progreso cada 100 generaciones
        if (gen + 1) % 100 == 0:
            print(f"Gen {gen+1:4d} | Mejor: {mejor_gen} | Promedio: {prom_gen:.1f}")

    print(f"\n Costo final óptimo encontrado: {mejor_costo_global}")
    return mejor_global, mejor_costo_global, historial_mejor, historial_promedio
```

```
#Test

random.seed(42)
np.random.seed(42)

mejor_sol, mejor_costo, hist_mejor, hist_prom = algoritmo_genetico(
    tam_poblacion=100,
    generaciones=500,
    prob_cruce=0.8,
    prob_mutacion=0.1,
    elitismo=2
)
```

El GA encontró su mejor solución muy rápido (antes de la generación 100) y se estabilizó, las 400 generaciones restantes no mejoras, debido a que el algoritmo ya convergió con las 100 primeras.

# Visualización de resultados

Se muestra la planificación final por día y la evolución del costo a lo larlo de las generaciones

```
def mostrar_solucion(solucion):
    """
    Imprime la asignación de tomas por día y el costo de cada día.
    """
    dias = {}
    for toma, dia in enumerate(solucion):
        if dia not in dias:
            dias[dia] = []
        dias[dia].append(toma + 1)

    print("\nPLANIFICACIÓN POR DÍA:")
    print("-" * 50)
    costo_total = 0
    for dia in sorted(dias.keys()):
        tomas_dia = dias[dia]
        actores_dia = set()
        for t in tomas_dia:
            for actor in range(N_ACTORES):
                if M[t-1][actor] == 1:
                    actores_dia.add(actor + 1)
        costo_dia = len(actores_dia)
        costo_total += costo_dia
        print(f"  Día {dia:2d}: Tomas {tomas_dia} → Actores {sorted(actores_dia)} → Costo: {costo_dia}")
    print("-" * 50)
    print(f"  COSTO TOTAL: {costo_total}")

def graficar_evolucion(historial_mejor, historial_promedio):
    plt.figure(figsize=(10, 5))
    plt.plot(historial_mejor,   label='Mejor costo', color='blue')
    plt.plot(historial_promedio, label='Costo promedio', color='orange', alpha=0.6)
    plt.xlabel('Generación')
    plt.ylabel('Costo')
    plt.title('Evolución del Algoritmo Genético — Problema de Doblaje')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```
random.seed(42)
np.random.seed(42)

# Corre el GA
mejor_sol, mejor_costo, hist_mejor, hist_prom = algoritmo_genetico(
    tam_poblacion=100,
    generaciones=500,
    prob_cruce=0.8,
    prob_mutacion=0.1,
    elitismo=2
)

# Muestra planificación por día
mostrar_solucion(mejor_sol)

# Muestra gráfica de evolución
graficar_evolucion(hist_mejor, hist_prom)
```

El GA encontró una planificación de 5 días con un costo total de 31 actor-días,
partiendo de soluciones aleatorias con costo promedio de 45. Esto representa una mejora del 24% respecto al punto de partida.

Todas las restricciones se cumplen: cada día tiene exactamente 6 tomas y se utilizan el mínimo de días posible (5).

El Día 2 es el más costoso con 7 actores convocados, mientras los demás días requieren solo 6.
